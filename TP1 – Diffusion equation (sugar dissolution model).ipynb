Le but de ce TP est de calculer numériquement de deux façons différentes la solution du problème suivant :

$$
    \tag{$\ast$}
    \begin{cases}
        \vphantom{\dfrac12}
        u_t(x,t) - u_{xx}(x,t) = 0, \qquad t > 0, \quad 0 < x < L, \\
        \vphantom{\dfrac12}
        u(x,0) = u_0(x), \\
        \vphantom{\dfrac12}
        u_x(0,t) = u_x(L,t) = 0. 
    \end{cases}
$$

Ce problème d'évolution peut modéliser, par exemple, la dissolution d'un sucre dans le café.
La valeur $u(x,t)$ représente alors la concentration en sucre (comprise entre $0$ et $1$) au point $x$ et à l'instant $t$.

La concentration initiale en sucre sera donnée par :

$$
    u_0(x) = 
    \begin{cases}
        1 \text{ si } x \in \left[ \dfrac {3L} 8, \dfrac {5L} 8 \right], \\
        \vphantom{\dfrac12}
        0 \text{ sinon,} 
    \end{cases}
$$

et on pourra choisir $L = 1$ dans les applications numériques.

**Question 1.** Justifier le fait de chercher $u$ sous la forme :

$$
    u(x,t) = \sum_{k=0}^{+\infty} c_k \exp \left( - \frac{k^2 \pi^2}{L^2} t \right) \cos \left( \frac{k \pi}{L} x \right).
$$

_Indication_: Appliquer la technique de séparation de variables et utiliser les conditions aux limites.

**Réponse 1.** Ecrivons $u(x,t)$ sous la forme $u(x,t)= f(x)g(t)$. Avec les conditions initiales, notre équation se réécrit :

\begin{equation*}
\begin{cases}
g'(t) f(x) = g(t) f''(t) \\
f(x)g(0) = u_0(x) \\
f'(0)g(t) = f'(L)g(t) = 0.
\end{cases}
\end{equation*}

Puisque que pour tous $t$, $g(t) \ne 0$, on a :

$$
f'(0)g(t) = 0 \Longleftrightarrow f'(0) = 0 \Longleftrightarrow f'(L)g(t) = 0 \Longleftrightarrow f'(L) = 0.
$$

On obtient ainsi :

$$ \frac{g'(t)}{g(t)} = c = \frac{f''(x)}{f(x)}$$ pour tous $t$ et $x$ où $g$ et $f$ ne s'annulent pas et où $c$ désigne une constante réelle.

D'une part, il vient :
$$
f''(x) = c f(x).
$$

Le polynôme caractéristique s'écrit :





D'aute part :
$$
g'(t) = c g(t)
$$

On a : $$ f(x)g(0) = u_0(x) $$ et $$ u_0,t) = f'(0)g(t) = f'(L)g(t) = 0 \Rightarrow f'(0) = f'(L) = 0 $$

On a $$ g'(t) f(x) = g(t) f''(t) $$ et d'après toute les formules précedente on a $$f''(x) - cf(x) = 0 $$ il s'agit d'une équation différentiel d'ordre 2.

1) Si $c = \lambda ² , \lambda > 0 $
$$ f(x) = Ae^{\lambda x} + Be^{\lambda x} \Rightarrow  f'(x) =  Ae^{\lambda x} - Be^{-\lambda x}$$
D'après les conditions initiale on a :
$$ f'(0) = 0 \Rightarrow A\lambda - B\lambda = 0 \Rightarrow A \lambda = B \lambda \Rightarrow A = B $$
$$ f'(L) = 0 \Rightarrow A\lambda e^{\lambda L} - B\lambda e^{-\lambda L} = 0 \Rightarrow  A\lambda e^{\lambda L} = B\lambda e^{-\lambda L} \Rightarrow A e^{\lambda L} = B e^{-\lambda L} \Rightarrow A e^{2 \lambda L} - B = 0 \Rightarrow A(e^{2 \lambda L } - 1 ) = 0 $$ avec $A = B$. On a donc comme $e^{2 \lambda L } - 1 = 0$.
$$ \Rightarrow A = B = 0 $$


2) Si $c = 0$ :
$$ f''(x) = 0 \Rightarrow f'(x) = A \Rightarrow f(x) = Ax + B \Rightarrow f(x) = A$$
$$ f'(0) = 0 \Rightarrow f(0) = A$$
$$ f'(L) = 0 \Rightarrow A = 0 \Rightarrow f(x) = B $$

3) Si $c= - \lambda² < 0, \lambda >0$
$$ f(x) = A \cos(\lambda x ) + B \sin(\lambda x) \rightarrow f'(x) = \lambda A \sin(\lambda x) + \lambda B \cos (\lambda x) $$
$$ f'(0) = 0 \Rightarrow \lambda B \cos(\lambda \times 0) = \lambda = 0 \Leftrightarrow B = 0 $$
$$ f'(L) = 0 \Rightarrow -\lambda A \sin ( \lambda L) + \lambda B \cos(\lambda L) = 0 \Leftrightarrow  \lambda A \sin ( \lambda L) = \lambda B \cos(\lambda L) \Leftrightarrow A \sin ( \lambda L) =  B \cos(\lambda L) \Leftrightarrow A \sin(\lambda L) = 0 \Rightarrow \lambda L = n \pi \Rightarrow \lambda = \dfrac{n \pi }{L} $$

Ainsi on a $ f(x) = A \cos(\dfrac{n\pi}{L} x)$

On a vu précedemment que $f'(x) = ck f(x)$, et $ k = - \lambda^2 \Rightarrow  k = -(\dfrac{n \pi}{L})^2 $. Ainsi on a  $  f(x) = c_1 \exp{i (\dfrac{n \pi}{L})}^2 $ avec $c_1$ une constante à déterminer.

Finalement on a $ u(x,t) = \sum_{i=1}^{\infty} c_i \exp(- \dfrac{i^2 \pi^2}{L^2} t) \cos(\dfrac{i \pi}{L} x)$.

**Question 4.** Pour approcher $u(x,t)$, on considère la série tronquée

$$
    u_N(x,t) = \sum_{i=0}^N c_i \exp \left( - \frac{i^2 \pi^2}{L^2} t \right) \cos \left( \frac{i \pi}{L} x \right).

Pour $N = 10$, $N = 100$ et $N = 1000$, tracer la solution approchée $u_N$ aux instants $t=10^{-6}$, $t=10^{-4}$, $t=10^{-2}$ et $t=1$. Que constate-t-on pour les petites valeurs de $t$ ? Pourquoi ?

```python
import numpy as np                  # librairie numpy
import matplotlib.pyplot as plt     # librairie matplotlib
plt.style.use('ggplot')             # optionnel : plus jolis graphes (c'est subjectif !)

def c_0():
    ## utiliser la question 2 pour que cette fonction retourne la valeur de c_0
    return 1/4

def c_k( k ):
    ## utiliser la question 2 pour que cette fonction retourne la valeur de c_k ici
    return 2/(np.pi*k)*(np.sin((5*k*np.pi)/8)-np.sin((3*k*np.pi)/8))

def u_N( x, t, N, L ):
    ## utiliser la question 1 pour que cette fonction retourne la valeur de u_N ici 
    S=c_0()
    for i in range(1,N):
        c=c_k(i)
        S+=c*np.exp(-(i**2*np.pi**2*t)/L**2)*np.cos((i*np.pi*x)/L)
    return S

L = 1        # L est la taille de l'intervalle
nx = 1000      # nx est le nombre de points de discrétisation pour évaluer u_N
x =np.linspace(0,1,nx)         # x est le vecteur qui contiendra les points où u_N sera évalué

##########################################################################################
# pas besoin de modifier le code ci-dessous, il sert simplement à afficher les résultats #
##########################################################################################

fig, ax = plt.subplots( 1, 3, figsize = ( 16, 6 ) )     # on crée 3 sous-figures disposées horizontalement

for i in range(3):      # on a 3 valeurs de N à prendre en compte
    N = 10**(i+1)       # on définit N = 10**(i+1) : pour 0 < i < 2, on a bien N = 10, 100 et 1000
    ax[i].set_title("N = {0}".format(N))    # on fait figurer le nombre de points dans le titre de la figure 
    for j in range(4):      # pour chaque valeur de N, on a 4 valeurs de t à prendre en compte
        t = 10**(2*j-6)     # on définit t = 10**(2*j-6) : pour 0 < j < 3, on a bien t = 1e-6, 1e-4, 1e-2 et 1
        ax[i].plot( x, u_N( x, t, N, L ),                 # on trace u_N en fonction de t
                    label="t = {0:1.0e}".format(t) )      # on met la valeur de t dans la légende

# on déplace la légende sous les figures
handles, labels = ax[-1].get_legend_handles_labels()
fig.legend( handles, labels, loc='lower center', ncol=4 );
```

**Question 5.1.** Dans un premier temps, écrire sous forme matricielle, puis implémenter, un schéma explicite aux différences finies pour résoudre l'équation $(\ast)$. Comparer les résultats avec la solution obtenue à la **Question 4.**

```python
def u_0( x, L ):
    ## cette fonction retourne la condition initiale u_0(x)
    ## on utilise la fonction logical_and de numpy :
    ## ceci crée un tableau avec des "True" (=1) quand 3*L/8 < x < 5*L/8, et des "False" (=0) sinon
    ## enfin, on multiplie ce tableau par "1." pour le convertir en float
    return 1. * np.logical_and( 3*L/8 < x, x < 5*L/8 ) 

def compute_dt_explicit( dx ):
    ## cette fonction calcule, en fonction du pas d'espace dx, le pas de temps dt qui permet d'assurer la stabilité
    dt =(dx**2)/2
    return dt

def build_explicit_matrix( nx, dx, dt ):

    ## cette fonction construit la matrice pour le schéma explicite
    
    subdiag_exp =-np.ones(nx-1)   # construire un vecteur de taille nx-1 contenant la sous-diagonale de la matrice
    diag_exp    =2*np.ones(nx)   # construire un vecteur de taille nx   contenant la      diagonale de la matrice
    supdiag_exp =-np.ones(nx-1)   # construire un vecteur de taille nx-1 contenant la  sur-diagonale de la matrice
    
    diag_exp[ 0] = 1   # penser aux conditions aux limites
    diag_exp[-1] = 1  # penser aux conditions aux limites

    # cette commande permet d'assembler la sous-diagonale, la diagonale et la sur-diagonale pour créer la matrice A
    A = np.diag( subdiag_exp, k = -1 ) + np.diag( diag_exp, k = 0 ) + np.diag( supdiag_exp, k = 1 )

    # finalement, on crée la matrice M_exp
    M_exp =np.eye(nx) - dt/dx**2 * A

    # on retourne la matrice ainsi créée
    return M_exp

def update_explicit( u_n, M_exp ):
    ## cette fonction retourne u_np1 (la valeur de u^{n+1}) en fonction de u_n (la valeur de u^n) et de la matrice M_exp
    u_np1 =np.dot(M_exp,u_n)
    return u_np1

nx =1000 # nombre de points
dx =L/nx            # taille du pas d'espace
x = np.linspace(0,1,nx)            # points de discrétisation

t =0             # temps initial
t_end =10**-2         # temps final
dt =compute_dt_explicit(dx)            # pas de temps (calculé avec la fonction compute_dt_explicit)

M_exp =build_explicit_matrix(nx,dx,dt)         # matrice pour le schéma explicite (calculée avec la fonction build_explicit_matrix)

u =u_0(x,L)            # condition initiale (calculée avec la fonction u_0)

while t < t_end:
    ## boucle en temps : on part de u = u0, et on fait évoluer la valeur de u en utilisant update_explicit
    u =update_explicit(u,M_exp) 
    t+=dt
# pas besoin de modifier le code ci-dessous, il sert simplement à afficher les résultats
plt.plot( x, u_N( x, t_end, 100, L ), label="u Fourier" )   # on affiche la solution de Fourier avec 100 modes
plt.plot( x, u, label="u FD" )                              # on affiche la solution du schéma
plt.legend()
plt.show();                                               # on affiche la légende
```
